```python
#unknown-class probes classified as known-class
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

proto='tls'
dataset_name='selfdeploy24'

# 读取数据
df = pd.read_csv('result_by_service/proj_'+proto+'_'+dataset_name+'_withindex.csv')

# Parse the vector string as a numeric array
def parse_vector(vec_str):
    return np.array([float(x.strip()) for x in vec_str.strip('[]').split(',')])

# Convert the feature vector into a new DataFrame
features_df = pd.DataFrame(df['proj'].apply(parse_vector).tolist(), columns=[f'F{i}' for i in range(128)])

# Create the t-SNE model and perform dimensionality reduction
tsne = TSNE(n_components=2, random_state=42)
principal_components = tsne.fit_transform(features_df)

pc_df = pd.DataFrame(principal_components, columns=['PC1', 'PC2'])
pc_df['y_true'] = df['y_true']
pc_df['ip_pred'] = df['ip_pred']

# Remove the data where y true=26 and ip pred=25
pc_df = pc_df[~(((pc_df['y_true'] == 26) | (pc_df['y_true'] == 25)| (pc_df['y_true'] == 27)) & (pc_df['ip_pred'] == 25))]

# stat y_true count, visualize top5 organizations
y_true_counts = Counter(pc_df['y_true'])
top_6_y_true, top_6_counts = zip(*y_true_counts.most_common(5))

total_count = len(pc_df)
top_6_total = sum(top_6_counts)


# =============================================================================
# rgb_colors = [
#    # (110, 134, 172),  
#     (224, 110, 110),
#     (152, 202, 139),
#     (126, 180, 198),
#     (173, 147, 184),
#     #(245, 200, 181),
#     (185, 113, 125)  
# ]
# =============================================================================

rgb_colors = [  #dns
   # (110, 134, 172),  
    (224, 110, 110),
    (152, 202, 139),
    (245, 200, 181),
    (126, 180, 198),
    (173, 147, 184),
    #
    (185, 113, 125)   # the last color for 'other' class
]


color_mapping = {y_true: tuple(np.array(color)/255.0) for y_true, color in zip(top_6_y_true, rgb_colors)}


color_mapping['other'] = tuple(np.array(rgb_colors[-1])/255.0)

print("RGB Color Mapping:", color_mapping)

# Create the color mapping of ip_pred
special_cases = pc_df[((pc_df['y_true'] == 26) | (pc_df['y_true'] == 25)| (pc_df['y_true'] == 27)) & (pc_df['ip_pred'] != 25)]

ip_pred_unique = special_cases['ip_pred'].unique()

# Assign colors to ip_pred based on the y_true value in color_mapping
ip_pred_color_mapping = {}
for ip_pred in ip_pred_unique:
    ip_pred_color_mapping[ip_pred] = color_mapping.get(ip_pred, color_mapping['other'])  


set_label =  ['adscore',
 'ahrefs',
 'arbor',
 'archive',
 'binaryedge',
 'criminalip',
 'cybergreen',
 'fofa',
 'internet_census',
 'internettl',
 'intrinsec',
 'ipip',
 'leakix',
 'onyphe',
 'quadmetrics',
 'quake',
 'rapid7',
 'rau',
 'shadowserver',
 'shodan',
 'stretchoid',
 'tum',
 'webRay',
 'x_threatbook',
 'zoomeye',
'driftnet',
 'unknown',
 'censys'  ]


def assign_color_and_marker(row):
    if (row['y_true'] == 26 or row['y_true'] == 25 or row['y_true'] == 27) and row['ip_pred'] != 25:
        return ip_pred_color_mapping.get(row['ip_pred'], color_mapping['other']), '*'  
    elif row['y_true'] in color_mapping:
        return color_mapping[row['y_true']], 'o'  
    else:
        return color_mapping['other'], 'o'  

pc_df[['color_category', 'marker']] = pc_df.apply(assign_color_and_marker, axis=1, result_type='expand')

legend_labels = []

for y_true, count in zip(top_6_y_true, top_6_counts):
    if y_true == 26 or y_true == 25 or y_true == 27:
        continue  
    color = color_mapping.get(y_true, color_mapping['other'])
    #label = f'label={set_label[y_true]} ({count}/{total_count})'
    label = f'Label={set_label[y_true]}'
    legend_labels.append((label, color, 'o'))
    
    label = f'Pred={set_label[y_true]}'
    legend_labels.append((label, color, '*'))

legend_labels.append(('Label=others', color_mapping['other'], 'o'))
legend_labels.append(('Pred=others', color_mapping['other'], '*'))


handles = [plt.Line2D([0], [0], marker=marker, color=color, label=label, markersize=8, linestyle='')
           for label, color, marker in legend_labels]


plt.figure(figsize=(6.1, 3.5), dpi=300)


dot_size = 40*1.7  
star_size = 100*1.7  

for marker_type, group_data in pc_df.groupby('marker'):
    sns.scatterplot(
        x='PC1',
        y='PC2',
        data=group_data,
        #edgecolors='none',
        color=None,  
        hue='color_category',
        palette={k: k for k in group_data['color_category'].unique()},  
        marker=marker_type,  
        alpha=0.8,
        s=star_size if marker_type == '*' else dot_size,  
        legend=False,  
        ax=plt.gca()  
    )

plt.legend(handles=handles, bbox_to_anchor=(1, 1), loc='upper left',prop={'size':13.5},handletextpad=0.2)
plt.xlabel('')
plt.ylabel('')
ax = plt.gca()
ax.set_xticklabels([])
ax.set_yticklabels([])
plt.tight_layout()
plt.savefig(f'fig/known_{proto}.pdf')
plt.show()


```

```python
#clustering on still unknown-class probes 
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

proto='tls'
dataset_name='selfdeploy24'

if(proto=='http'):
    n_clusters = 6
if(proto=='tls'):
    n_clusters = 8
if(proto=='dns'):
    n_clusters = 3
    
df = pd.read_csv('result_by_service/proj_still_unknown_'+proto+'_'+dataset_name+'_withindex.csv')
def parse_vector(vec_str):
    return np.array([float(x.strip()) for x in vec_str.strip('[]').split(',')])

# Convert the feature vector into a new DataFrame
features_df = pd.DataFrame(df['proj'].apply(parse_vector).tolist(), columns=[f'F{i}' for i in range(128)])

# t-SNE model
tsne = TSNE(n_components=2, random_state=42)
principal_components_all = tsne.fit_transform(features_df)


pc_df_all = pd.DataFrame(data=principal_components_all, columns=['PC1', 'PC2'])
pc_df_all['y_true'] = df['y_true']
pc_df_all['ip_pred'] = df['ip_pred']

filtered_pc_df = pc_df_all[ ((pc_df_all['y_true'] == 26) | (pc_df_all['y_true'] == 25)| (pc_df_all['y_true'] == 27)) & (pc_df_all['ip_pred'] == 25)]

features_df = features_df[ ((pc_df_all['y_true'] == 26) | (pc_df_all['y_true'] == 25)| (pc_df_all['y_true'] == 27)) & (pc_df_all['ip_pred'] == 25)]

# Apply K-means clustering
kmeans = KMeans(n_clusters=n_clusters, init='k-means++', max_iter=300, n_init=10, random_state=42)
kmeans.fit(features_df)
labels = kmeans.labels_

filtered_pc_df = filtered_pc_df.assign(Cluster=labels)

rgb_colors = [
    (110, 134, 172),
    #(224, 110, 110),
    (152, 202, 139),
    (126, 180, 198),
    (173, 147, 184),
    (245, 200, 181),
    
    (253, 211, 99),
    (185, 113, 125),
    (254, 144, 21)
]
color_palette = [tuple(np.array(color) / 255.0) for color in rgb_colors]

plt.figure(figsize=(5.1, 3.5),dpi=300)

for cluster in range(n_clusters):
    cluster_data = filtered_pc_df[filtered_pc_df['Cluster'] == cluster]
    plt.scatter(
        cluster_data['PC1'],
        cluster_data['PC2'],
        color=color_palette[cluster],  
        alpha=0.7,
        s=50,  
        label=f'Cluster {cluster}'
    )

plt.legend(bbox_to_anchor=(1, 1), loc='upper left',prop={'size':13},handletextpad=0.2)
plt.xlabel('')
plt.ylabel('')
ax = plt.gca()
ax.set_xticklabels([])
ax.set_yticklabels([])
plt.tight_layout()  
plt.savefig(f'fig/still_unknown_{proto}.pdf')
plt.show()

temp_df = filtered_pc_df.copy()
temp_df['ClusterNo'] = labels
df_indices = df[((df['y_true'] == 26) | (df['y_true'] == 25)| (df['y_true'] == 27)) & (df['ip_pred'] == 25)].index
df.loc[df_indices, 'ClusterNo'] = temp_df.loc[temp_df.index, 'ClusterNo'].values

df['ClusterNo'] = df['ClusterNo'].fillna(-1)

# Save dataframe to CSV file
df.to_csv('result_by_service/proj_still_unknown_'+proto+'_'+dataset_name+'_with_clusterid.csv', index=False)
```
